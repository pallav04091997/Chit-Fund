```python
import numpy as np
import pandas as pd
```

```python
chit = pd.read_csv("chit_fund_exercise.csv")
```

```python
chit.head()
```

```python
def A():
    sum_paid = 0
    for i,j in enumerate(chit.Month):
        sum_paid= sum_paid + chit.Contribution[i] - chit.Amount_returned_to_everyone_in_the_group[i]
        
    return sum_paid
A()
#So in the end the last person total Contribution is 43,800 and the pot he gets was of 47,500
#so in total he makes 3,700
        
```

```python
chit['bid'] = 50000-chit['Amount_won_by_the_bidder']
```

```python
chit
```

```python
def B():
    sum2 = 0
    for i,j in enumerate(chit.Month):
        sum2 = sum2+ chit.Contribution[i] - chit.Amount_returned_to_everyone_in_the_group[i]
        
    sum2 =sum2+ chit.bid[0]
    return chit.Net_amount_recd_by_Bid_winner[0]-sum2 
B()
```

```python
amount_invested = []
for i,j in enumerate(chit.Month):
    a=0
    a = 50000 + chit.bid[i] - chit.Amount_returned_to_everyone_in_the_group.sum()
    amount_invested.append(a)
    
```

```python
amount_invested
```

```python
d = {}
for i,j in enumerate(chit.Month):
    d['bidder '+ str(i+1)] = chit.Net_amount_recd_by_Bid_winner[i] - amount_invested[i]
```

```python
d
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(range(len(d)), list(d.values()))
plt.xticks(range(len(d)), list(d.keys()), rotation = 'vertical')
plt.title('profit vs loss on total investment including bidding amount(₹)')
```

```python
per = {}
for i,j in enumerate(chit.Month):
    per['bidder '+str(i+1)] = (d['bidder '+str(i+1)] / amount_invested[i])*100
```

```python
per
```

```python
%matplotlib inline
plt.bar(range(len(per)), list(per.values()))
plt.xticks(range(len(per)), list(per.keys()), rotation = 'vertical')
```
